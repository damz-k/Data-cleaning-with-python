# Assigment

- rename the columns and make them more descriptive
- change Pclass to first, second and third class.
- split the name column to last name and other names
- make the age an integer
- delete the ticket and cabin columns
- make the embarked column more descriptive
- search for 10 names and correct ages of the null age and replace it

# solution

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Titanic Data.csv")
```

```python
df
```

# 1

```python
df.rename(columns={'Pclass': 'Position', 'Ticket': 'TicketID', 'Cabin': 'CabinID'}, inplace=True)
```

# 2

```python
df['Position'] = df['Position'].replace({3: 'Third Class', 2: 'Second Class', 1: 'First Class'})
```

```python
df.head(5)
```

# 3

```python
df[['Name', 'LastName']] = df['Name'].str.split(' ', n=1, expand=True)
```

```python
df = df[['PassengerId', 'Position', 'Name', 'LastName', 'Sex', 'Age', 'SibSp', 'Parch', 'TicketID', 'Fare','CabinID', 'Embarked', 'Survived']]
```

```python
df.head(5)
```

# 4

```python
def modify_Sex(value):
    if value == 'male':
        return '1'
    elif value == 'female':
        return '0'
        return value
```

```python
df['Sex'] = df['Sex'].apply(modify_Sex)
```

```python
df.head(5)
```

# 5

```python
df.drop(columns=['TicketID', 'CabinID'], inplace=True)
```

# 6

```python
def modify_embarked(value):
    if value == 'S':
        return 'Southampton'
    elif value == 'C':
        return 'Cherbourg'
    elif value == 'Q':
        return 'Queenstown'
    return value
```

```python
df['Embarked'] = df['Embarked'].apply(modify_embarked)
```

```python
df.head(5)
```

# 7

```python
df[df['Age'].isna()][['Name', 'Age']]
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'] = df['Age'].fillna(0).astype(int)
```
